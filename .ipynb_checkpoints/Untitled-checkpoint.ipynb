```python
#import numpy as np
#import tensorflow as tf
import os
from skimage import io, transform
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline


def load_data(data_directory):
    #Fait une liste des dossiers.
    directories = [d for d in os.listdir(data_directory)
                   if os.path.isdir(os.path.join(data_directory, d))]
    images = []
    labels = []
    # loop sur les dossiers (labels)
    for d in directories:
        label_directory = os.path.join(data_directory,d) #donne le path des labels
        # fait une lite des fichiers dans un dossier (labels)
        file_names = [os.path.join(label_directory, file) for file
                      in os.listdir(label_directory) if file.endswith(".ppm")]
        # la on a toutes les images d'un dossier, loop sur eux
        for file in file_names:
            images.append(io.imread(file))
            labels.append(int(d))
    return images, labels


Train_images, Train_labels = load_data("/Users/simonboivin/Documents/BelgiumTSC/Training")
Test_images, Test_labels = load_data("/Users/simonboivin/Documents/BelgiumTSC/Testing")
```

```python
def one_hot_encode(vec, vals=62):
    n=len(vec)
    out = np.zeros((n,vals))
    out[range(n),vec]=1
    return out
```

```python
Train_labels=one_hot_encode(Train_labels)
```

```python
Train_images28.shape
```

```python
class Batch():
    
    def __init__(self,images,labels):
        self.i = 0
        self.images = images
        self.labels = labels
    
    def next_batch(self, batch_size):
        # Note that the 100 dimension in the reshape call is set by an assumed batch size of 100
        x = self.images[self.i:self.i+batch_size].reshape(batch_size,28,28,3)
        y = self.labels[self.i:self.i+batch_size]
        self.i = (self.i + batch_size) % len(self.images)
        return x, y
```

```python
data = Batch(Train_images28,Train_labels)
```

```python
data.next_batch(100)
```
